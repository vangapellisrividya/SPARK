```python

from pyspark.sql import SparkSession,types
from pyspark import SparkConf
from pyspark.sql.functions import *
spark = SparkSession.builder.getOrCreate()
```

```python
'''To read from json file from local'''
df = spark.read.json("file.json",multiLine=True)
df.printSchema()
```

```python
'''To write to the json file in hdfs'''
df.write.json("hdfs://localhost:9000/json/writeto.json")
df.show(2)
```

```python
df1 = spark.read.csv("CpuLogData2019.csv",header=True)
df1.show(5)
```

```python
'''Write to csv file in hdfs'''
df1.write.option("header","true")\
.csv("hdfs://localhost:9000/jsonread/writeto.csv")
```
